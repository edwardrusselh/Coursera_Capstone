<h4> Identifying Cuisines or Restaurants in the Neighborhood </h4>

Putting up a restuarant require some studies. For one, when setting up a restaurant, it is essential to look into the current set-up of the area or the city. It may help to look into the available restaurants in the area. Look into the trends, and what are the avaiable cuisines in the city. Setting up a restaurant that introduces a new cuisine can be good. However, this can be risky at times. People of a certain community has a certain preference and this is that this project will look into. 

The aim of this project is to look into the available cuisines or restaurant in the area of <b> Central Toronto </b>. This will help determine business persons know what type of restaurants would people prefer in a certain community. <b> Do people prefer Chinese, French, Italian, Japanese or others? Which among the neighborhood (of Central Toronto) prefers sushi restaurants, for example? </b>

Hopefully, this project will be able to advise the business person to determine what people from Central Toronto prefer by looking the avaialble restaurant in the neighborhoods. Also, this may look at the trend of each neighborhood by looking at the current cuisines avaiable in the area.
